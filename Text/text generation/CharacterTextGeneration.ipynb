```python
# character text generation can sometimes be even better then word generation(idk why, maybe because of the high cardinality)
#thus keep this in mind
import tensorflow as tf
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.preprocessing.sequence import pad_sequences

import numpy as np
import os
import time
```

```python
path_to_file = tf.keras.utils.get_file('shakespeare.txt', 'https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt')
path_to_file
```

```python
text = open(path_to_file, 'rb').read().decode(encoding='utf-8')
all_chars = sorted(set(text))
all_chars
```

```python
sl = preprocessing.StringLookup(vocabulary=list(all_chars), mask_token=None)
to_numeral = sl(tf.strings.unicode_split(text, input_encoding='UTF-8'))
```

```python
sl.get_vocabulary()
```

```python
texts = []
labels = []
seq_len = 100

for i in range(5000-seq_len):
  section = to_numeral[i: i+seq_len]
  for j in range(2, len(section)):
    subsection = section[:j]
    texts.append(subsection[:-1].numpy())
    labels.append(subsection[-1].numpy())

texts, labels
```

```python
texts, labels = np.array(texts), np.array(labels)
texts = pad_sequences(texts, maxlen=seq_len)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Embedding(len(all_chars)+1, 64, input_length=seq_len),
  tf.keras.layers.Conv1D(64, 5, activation="relu"),
  tf.keras.layers.MaxPool1D(),
  tf.keras.layers.Bidirectional(tf.keras.layers.GRU(64)),
  tf.keras.layers.Dense(len(all_chars)+1, activation="softmax")
])
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam")
model.fit(texts[:-1], labels, epochs=5)
```

```python
def give_text(query = "S", num_chars=500):
  for i in range(num_chars):
    to_pred = sl(tf.strings.unicode_split(query, input_encoding='UTF-8'))
    padded = pad_sequences(np.array([to_pred]), maxlen=seq_len)
    max_index = np.argmax(model.predict(padded)[0])
    to_text = preprocessing.StringLookup(vocabulary=list(all_chars), mask_token=None, invert=True)
    char = to_text(np.array([max_index]))[0]
    char = char.numpy()
    # print(char)
    print(i)
    query += char.decode("utf-8") 
  print(query)


give_text("ABHIRAAM:")
```
